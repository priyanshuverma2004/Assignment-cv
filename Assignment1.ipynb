```python
def comparecenter(img,centre,x,y):
  val = 0
  try:
    if img[x][y]>=centre:
      val = 1
  except:
      pass
  return val

def LBPcal(img,x,y):
  centre = img[x][y]
  b = []
  b.append(comparecenter(img,centre,x-1,y-1)) #for top left
  b.append(comparecenter(img,centre,x-1,y))   #for top
  b.append(comparecenter(img,centre,x-1,y+1)) #for top right
  b.append(comparecenter(img,centre,x,y-1))   #for left
  b.append(comparecenter(img,centre,x,y+1))   #for right
  b.append(comparecenter(img,centre,x+1,y-1)) #for bottom left
  b.append(comparecenter(img,centre,x+1,y))   #for bottom
  b.append(comparecenter(img,centre,x+1,y+1)) #for bottom right

  val = 0
  for i in range(len(b)):
    val+=b[i]*2**i

  return val

def LBP(img_gray):
  height, width = img_gray.shape
  img_lbp = np.zeros((height,width),np.uint8)
  for i in range(height):
    for j in range(width):
      img_lbp[i][j] =  LBPcal(img_gray,i,j)
  return img_lbp
```

```python
lbpimg1=LBP(gimg1)
lbpimg2=LBP(gimg2)
print("LBP image 1:")
cv2_imshow(lbpimg1)
print("LBP image 2:")
cv2_imshow(lbpimg2)
```
